### Usando o GPT4o (gpt-4o-2024-08-06)

```python
import time

MAX_RETRIES = 30
RETRY_DELAY_SECONDS = 2

global_delay_errors = 0

for target in targetInvest:
    for news in csv['content']:
        payload = {
            "text": news,
            "investmentTarget": target
        }

        retries = 0
        while retries <= MAX_RETRIES:
            response = requests.post(
                url,
                data=json.dumps(payload),
                headers=headers
            )

            if response.status_code == 200:
                print("Status: 200")
                break

            if response.status_code == 429:
                retries += 1
                global_delay_errors += 1

                print(
                    f"429 rate limit — retry {retries}/{MAX_RETRIES} "
                    f"(total delays: {global_delay_errors}), "
                    f"aguardando {RETRY_DELAY_SECONDS}s"
                )

                time.sleep(RETRY_DELAY_SECONDS)
                continue

            print(f"Erro HTTP {response.status_code}")
            break

print(
    f"\nOcorreram {global_delay_errors} erros de rate limit (HTTP 429). "
    f"Isto adicionou aproximadamente "
    f"{global_delay_errors * RETRY_DELAY_SECONDS}s ao tempo total de execução."
)
```

## Avaliação dos modelos
Comparação com o dataset gabarito gerado

```python
csv_ararafin_evaluate = pd.read_csv("https://raw.githubusercontent.com/vinelouzada/ararafin-research/refs/heads/main/ararafin_llm_avaliacao.csv")
```

```python
def evaluate_classification(y_true, y_pred):
    return {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, average="weighted"),
        "f1": f1_score(y_true, y_pred, average="weighted"),
    }

def print_metrics(metrics):
    print(f"Acurácia: {metrics['accuracy']:.2f}")
    print(f"Precisão: {metrics['precision']:.2f}")
    print(f"F1-score: {metrics['f1']:.2f}")

TARGETS = {
    "FIIS": "Sentimento esperado para FIIs",
    "RENDA_FIXA": "Sentimento esperado para Renda Fixa",
    "ACAO": "Sentimento esperado para Ações",
}
```

```python
def evaluate_model(df_model, df_gabarito):
    df = pd.merge(
        df_model,
        df_gabarito,
        left_on="text",
        right_on="content"
    )

    results = {}

    for target, expected_col in TARGETS.items():
        subset = df[df["investment_target"] == target]

        metrics = evaluate_classification(
            subset[expected_col],
            subset["classification"]
        )

        results[target] = metrics

        print(f"\n {target}")
        print_metrics(metrics)

    return results
```

```python
results_gpt_5_1 = evaluate_model(gpt_5_1_df, dataset_gabarito)
```

## Análise de Tempo de Execução dos Modelos

Esta seção apresenta a análise do tempo total de execução dos modelos avaliados, considerando aspectos de eficiência computacional e limitações operacionais observadas durante o experimento. Para melhor organização, os modelos foram agrupados por família.

---

### Família GPT (OpenAI)

| Modelo        | Tempo de Execução | Observações                                              |
| ------------- | ----------------- | -------------------------------------------------------- |
| **gpt5.1**    | 95m 14s       | Tempo total de execução do experimento.                  |
| **gpt5-mini** | 253m 1.7s         | Maior tempo de execução entre os modelos avaliados.      |
| **gpt-4o**    | 57m 51.5s         | Execução impactada por erros de *rate limit* (HTTP 429). |

**Observação técnica:**  
Durante a execução do **gpt-4o**, foi necessário implementar uma estratégia de *retry* com intervalo fixo de **2 segundos** entre tentativas. Essa medida foi adotada devido ao **baixo limite de tokens por minuto imposto pelo modelo**, uma vez que a taxa de requisições da aplicação excedia esse limite.  

Ao todo, ocorreram **122 erros de rate limit (HTTP 429)**, adicionando aproximadamente **244 segundos** ao tempo total de execução. Sem esse impacto, o tempo estimado de execução do gpt-4o seria de cerca de **53m 47s**.

---

### Família Maritaca AI

| Modelo        | Tempo de Execução | Observações |
|--------------|-------------------|-------------|
| **sabiazinho** | 32m 52.6s | Menor tempo de execução do experimento. |
| **sabia 3.1** | 80m 42.3s | Tempo intermediário, sem interferências externas registradas. |

---

### Comparação Geral

| Família | Modelo mais rápido | Modelo mais lento |
|-------|-------------------|-------------------|
| GPT | gpt-4o | gpt5-mini |
| Maritaca AI | sabiazinho | sabia 3.1 |

---

### Síntese

- Os modelos da **família Maritaca AI** apresentaram menor tempo médio de execução, com destaque para o *sabiazinho*.
- O **gpt-4o** demonstrou bom desempenho computacional, porém com sensibilidade a restrições operacionais, especialmente relacionadas ao limite de tokens por minuto.
- O **gpt5-mini** apresentou o maior custo temporal, o que pode impactar sua adoção em cenários de maior escala.

Os resultados evidenciam que limitações de infraestrutura e políticas de uso dos modelos podem influenciar significativamente o tempo de execução, devendo ser consideradas na avaliação comparativa entre diferentes soluções de linguagem natural.

### Análise das métricas de desempenho

```python
def plot_metric_heatmap(values, modelos, dominios, metric_name):
    plt.figure(figsize=(8,4))
    plt.imshow(
        values,
        aspect='auto',
        cmap='RdYlGn',
        vmin=0.1,
        vmax=1.0
    )

    plt.xticks(range(len(dominios)), dominios)
    plt.yticks(range(len(modelos)), modelos)

    for i in range(len(modelos)):
        for j in range(len(dominios)):
            plt.text(
                j, i,
                f'{values[i, j]:.2f}',
                ha='center',
                va='center',
                fontsize=11,
                color='black'
            )

    plt.colorbar(label=metric_name)
    plt.tight_layout()
    plt.show()
```

```python
plot_metric_heatmap(acuracia, llms, dominios, 'Acurácia')
plot_metric_heatmap(precisao, llms, dominios, 'Precisão')
plot_metric_heatmap(f1_score, llms, dominios, 'F1-score')
```
